```python
#FATHIMA IRFANA
#Roll no: 30
#Aim:
#1)a)Load iris dataset into a dataframe
import pandas as pd
from sklearn.datasets import load_iris

# a) Load iris dataset into a dataframe
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# b) Print shape of dataset, display all columns and their data types
print("Shape of dataset:", df.shape)
print("\nColumns and data types:\n", df.dtypes)

# c) Take descriptive statistics
print("\nDescriptive Statistics:\n", df.describe())

# d) Delete all rows with null values
df = df.dropna()

# e) Select the Species classes, count the number of examples in each class
print("\nSpecies class distribution:\n", df['species'].value_counts())

# f) Map the species classes into 1, 2, and 3
species_mapping = {'setosa': 1, 'versicolor': 2, 'virginica': 3}
df['species_mapped'] = df['species'].map(species_mapping)

# g) Print the standard deviation and mean of Sepal length
print("\nSepal Length Mean:", df['sepal length (cm)'].mean())
print("Sepal Length Std Dev:", df['sepal length (cm)'].std())
```

```python
#2)Create a DataFrame with 4 records having the fields:
#StudentID, RegNo, FullName, Gender, Department, Marks  and Perform the following operations:
import pandas as pd
import numpy as np
data = {
    'StudentID': [201, 202, 203, 204],
    'RegNo': ['R101', 'R102', 'R103', 'R104'],
    'FullName': ['Asha', 'Binu', 'Chitra', 'Deepak'],
    'Gender': ['F', 'M', 'F', 'M'],
    'Department': ['CSE', 'ECE', 'IT', 'ME'],
    'Marks': [87, np.nan, 78, 92]  # one student has a missing mark
}
#a) Insert a new column 'Grade'  into the DataFrame and assign values.
df = pd.DataFrame(data)
df['Grade'] = ['A', 'B', 'C', 'A']
print(df)
#b) Find and display the total no.of rows  in the DataFrame.
print("Total rows:", len(df))
#c) Select the first two rows and the last two rows of the DataFrame.
print("\nFirst two rows:\n", df.head(2))
print("\nLast two rows:\n", df.tail(2))
#d) Drop all rows where any column has a null (NaN) value.
df = df.dropna()
print("\nDataFrame after dropping rows with null values:\n", df)
#e)Remove the 'RegNo column from the DataFrame.
df = df.drop(columns=['RegNo'])
print("\nDataFrame after removing 'RegNo' column:\n", df)
#f) Display the 'FullName' and 'Marks' from rows 3 to 6.
print("\n'FullName' and 'Marks' from rows 3 to 6:\n", df.loc[2:5, ['FullName', 'Marks']])
#g) Replace all null values in 'Marks' column with the median of that column
median_marks = df['Marks'].median()
df['Marks'] = df['Marks'].fillna(median_marks)
print("\nDataFrame after replacing null values in 'Marks' with median:\n", df)
```
