```python
#FATHIMA IRFANA N P
#ROLL NO : 30
#AIM:
#Write a python program  to implement iris classification(Use only two features and two classes of species ) using SVM classifier.
from sklearn.preprocessing import StandardScaler
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
iris=pd.read_csv("iris.csv")
iris.info()
x=iris.iloc[0:100,0:2]
y=iris.iloc[0:100,-1]
print(x)
print(y)
```

```python
scaler=StandardScaler()
x_std=scaler.fit_transform(x)

```

```python
from sklearn.svm import SVC
svc=SVC(C=1.0,kernel='linear',degree=3, gamma='auto')
model=svc.fit(x_std,y)
```

```python
color=['black' if c=='Setosa' else 'red' for c in y]
plt.scatter(x_std[:,0],x_std[:,1],c=color)
```

```python
w=svc.coef_
print(w)
```

```python
color = ['black' if c== 0 else 'red' for c in y]
plt.scatter(x_std[:,0],x_std[:,1],c=color)
w=svc.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(-2.5,2.5)
yy=a*xx-(svc.intercept_[0])/w[1]
plt.plot(xx,yy)
plt.axis("off"),plt.show();
```
