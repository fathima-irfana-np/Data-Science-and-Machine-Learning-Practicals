```python
#Write a numpy program to create a 8 x 8 matrix and fill it with a checkerboard pattern
# [[1 1 0 0 0 0 1 1]
# [0 0 1 1 1 1 0 0]
# ..
# â€¦
# [0 0 1 1 1 1 0 0]]

matrix=np.ones((8,8),dtype=int)
matrix[::2,2:-2]=0
matrix[1::2,:2]=0
matrix[1::2,-2:]=0
print(matrix)
```

```python
# 9)Write a numpy program to compute
#    a) multiplication of two given matrices.
#    b) addition of two matrices

array1=np.array([[7,8,9],[4,5,6],[1,2,3]])
array2=np.array([[17,18,19],[14,15,16],[11,12,13]])
mul=np.dot(array1,array2)
add=array1+array2
print("mul=\n",mul)
print("add=\n",add)
multi=np.multiply(array1,array2)
print("mul(like add)=\n",multi)
multi=np.matmul(array1,array2)
print("matmul(like dot)=\n",multi)
```

```python
# 10)Write a numpy program to eigenvalues and eigenvectors of a given matrix
array1=np.array([[2,2,1],[1,3,1],[1,2,2]])
w , v= np.linalg.eig(array1)
print("eigen value:\n", w)
print("eigen vector:\n", v)
```

```python
# 11)Write a program to find transpose, determinant and inverse of a matrix.
array1=np.array([[2,2,1],[1,3,1],[1,2,2]])

print("\nmatrix:\n", array1)

print("\ntranspose of matrix:\n", np.transpose(array1))

determinant_result=np.linalg.det(array1)
print("\ndetereminent of matrix:\n", round(determinant_result, 6))

print("\ninverse of matrix:\n", np.linalg.inv(array1))
```
