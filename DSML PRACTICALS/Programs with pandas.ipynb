```python
# 3..Write a Pandas program to create and display a DataFrame ‘Examdata’ from a specified dictionary data(5 records) which has the fields 1)name 2)Score 3)Attempts 4)Qualify

import numpy as np
examdata = {'name': ['A', 'B', 'C'],
'score': [17.5, 10, np.nan],
'attempts': [1, 3, 2],
'qualify': ['yes', 'no', 'no']}
df = pd.DataFrame(examdata)
print(" Original DataFrame 'Examdata' == ")
print(df)
print()

# b)save it as a csv file and retrieve it.
filename = 'df.csv'
df.to_csv(filename, index=False)
print(f"Data frame saved to the '{filename}'")
retrieved_data=pd.read_csv(filename)
print()
print(f"Data retrieved from the '{filename}'")
print(retrieved_data)
print()

# c)Select 'name' and 'Qualify' columns in rows 1, 2, 3
dff = retrieved_data.copy()
select_columns = dff.loc[[0, 1, 2], ['name', 'qualify']]
print(select_columns)
print()

# d)select the rows where the number of attempts in the examination is greater than 1
print(" Select rows where attempts > 1 ")
attempts_g1 = dff[dff['attempts'] > 1]
print(attempts_g1)
print()
;/
# e)count the number of rows and columns of the DataFrame
print("Count rows and columns ")
rowno, colsno = df.shape
print(f"Number of rows: {rowno}")
print(f"Number of columns: {colsno}")
print()

# f)select rows where Number of attempts in the examination is less than 3 and score greater than 5
print("Select rows where attempts < 3 AND score > 5 ")
filter_row = df[(df['attempts'] < 3) & (df['score'] > 5)]
print(filter_row)
print()

# g)select rows where score are missing
print("select rows where score are missing")
missing_score = dff[dff['score'].isnull()]
print(missing_score)
print("")


# h)change the score in row 2 to 12
print("change the score in row 2 to 12")
dff.loc[1, 'score'] = 12
print(dff)
print("")


# i)delete the last row ,append a new row and return the original data frame
print("Delete last row, append new row ")
# Delete the last row index 4
after_deletion = dff.drop(dff.index[-1])
print("DataFrame after deleting the last row:")
print(after_deletion)
print("")
# Append new row
new_row = {'name': 'G', 'score': 20, 'attempts': 1, 'qualify': 'yes'}
after_append = pd.concat([after_deletion, pd.DataFrame([new_row])])
print("\nDataFrame after appending a new row:")
print(after_append)

dff = after_append.copy()
print("")

# j)replace the 'qualify' column contain the values 'yes' and 'no' with True and False.
print("Replace 'qualify' values with True/False ")
dff['qualify'] = dff['qualify'].map({'yes': True, 'no': False})
print(dff)
print("")


# k)insert a new column ‘Rank’ and populate it.
print("insert a new column ‘Rank’ and populate it.")
# dff['Rank'] = range(0, len(dff))
dff['Rank'] = [2 , 3 ,1]
print(dff)
print("")
```
