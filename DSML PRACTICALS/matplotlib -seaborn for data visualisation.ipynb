```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)


# Display first few rows
print(df.head())

# Line Plot
df.plot(kind = 'line')
df.plot(kind = 'line',subplots=True)
df["sepal length (cm)"].plot(kind = 'line')
plt.show()
```

```python
# Histograms
df.plot(kind='hist', bins=20, alpha=0.6, edgecolor="black", figsize=(10,6), title="Histogram - All Features")
plt.show()
```

```python
# Histogram for one feature
x = df["sepal length (cm)"]
plt.hist(x, bins=20, color="green", edgecolor="black")

plt.title("Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.show()
```

```python
# Boxplot (Sepal Width)

plt.figure(figsize=(6,5))
plt.boxplot(df['sepal width (cm)'])
plt.title("Boxplot - Sepal Width")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.boxplot([df[col] for col in df.columns], labels=df.columns)
plt.title("Boxplot - All Features")
plt.ylabel("cm (values)")
plt.xticks()
plt.show()
```

```python
# Bar Chart (Mean Petal Length per Species)
df['species'] = iris.target_names[iris.target]
species_mean = df.groupby('species')["sepal length (cm)"].mean()

plt.figure(figsize=(6,5))
plt.bar(species_mean.index, species_mean.values, color=['red','green','blue'])
plt.title("Barplot - Mean Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Mean Sepal Length (cm)")
plt.show()
```

```python
plt.bar(
    df['species'].unique(),
    df.groupby('species')['sepal length (cm)'].mean(),
    color=['red', 'green', 'blue']
)

plt.title("Mean Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Mean Sepal Length (cm)")
plt.show()
```

```python
#Barchart
x=df["sepal length (cm)"]
plt.bar(df.index,x)
plt.title('Bar Chart of Sepal Length')
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.show()
```

```python

# Pie Chart (Species Distribution)

plt.figure(figsize=(6,6))
species_counts = df['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart - Species Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

species_counts = df['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title("Species Distribution")
plt.show()
```

```python
#b)Write a program to draw multivariate visualization plots(scatter plots, scatter multiple,scatter matrix,bubble plot) with matplotlib for iris dataset

from pandas.plotting import scatter_matrix

# 1. Scatter Plot (bivariate)
plt.figure(figsize=(6,5))
df.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)')
plt.title("Scatter Plot - Sepal Length vs Sepal Width")
plt.show()

# 2. Multiple Scatter Plots (overlayed)
# Multiple scatter plots overlayed
ax1 = df.plot(kind='scatter', x='petal length (cm)', y='petal width (cm)',
              label='Petal Length vs Petal Width', color='r')

ax2 = df.plot(kind='scatter', x='petal length (cm)', y='sepal length (cm)',
              label='Petal Length vs Sepal Length', color='g', ax=ax1)
ax3 = df.plot(kind='scatter', x='petal length (cm)', y='sepal width (cm)',
              label='Petal Length vs Sepal Width', color='b', ax=ax2)
ax4 = df.plot(kind='scatter', x='sepal length (cm)', y='sepal width (cm)',
              label='Sepal Length vs Sepal Width', color='orange', ax=ax3)

plt.title("Multiple Scatter Plots")
plt.show()

```

```python

# 3. Scatter Matrix
plt.figure(figsize=(10,10))
scatter_matrix(df.iloc[:, :4], alpha=0.5, figsize=(10,10), diagonal='kde')
plt.suptitle("Scatter Matrix - All Features", fontsize=16)
plt.show()

```

```python

# Multivariate Bubble Plot (4 features)
plt.figure(figsize=(8,6))

bubble = plt.scatter(
    df['sepal length (cm)'],      # x-axis
    df['sepal width (cm)'],       # y-axis
    s=df['petal length (cm)']*30, # bubble size
    c=df['petal width (cm)'],     # bubble color
    cmap='viridis',               # colormap
    alpha=0.6,
    edgecolors="black"
)

plt.title("Multivariate Bubble Plot - 4 Features")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

# Add colorbar for petal width
cbar = plt.colorbar(bubble)
cbar.set_label("Petal Width (cm)")

plt.show()
```

```python
# 4. Bubble Plot

plt.figure(figsize=(7,6))
plt.scatter(
    df['petal length (cm)'],
    df['petal width (cm)'],
    s=df['sepal length (cm)']*20,   # bubble size scaled by sepal length
    alpha=0.5,
    c='blue'
)
plt.title("Bubble Plot - Petal Length vs Petal Width (Bubble = Sepal Length)")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
#c)Write a program to draw univariate and multivariate visualization plots with seaborn(line plot, pairplot,jointplot,heatmap) for iris dataset
import seaborn as sns

#  Pairplot (Multivariate)
sns.pairplot(df, hue="species", kind="scatter", diag_kind="kde")
plt.suptitle("Pairplot - All Features by Species", y=1.02)
plt.show()



```

```python
# Jointplot (Bivariate + Univariate)

sns.jointplot(x="sepal length (cm)", y="sepal width (cm)", data=df, hue="species", height=6)
plt.show()
```

```python
# Heatmap (Correlation Matrix)

plt.figure(figsize=(8,6))
sns.heatmap(df.iloc[:, :4].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap - Correlation Between Features")
plt.show()
```
